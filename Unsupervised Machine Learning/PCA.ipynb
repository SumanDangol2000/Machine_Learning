unsupervised tries to find the relations within a features itself not with target value like supervised ML.

1. Centring (take mean of feature, subtract each features data with mean)
2. Calculate the covariance matrix
3. Calculate eigen values
4. Calculate eigen vectors


```python
#Import python libraries
import numpy as np
```

```python
#Create a class PCA along with methods to predict hte result
class PCA:
  def __init__(self):
    self.num_components = 1
    self.mean = None

  def fit_transform(self, x):
      self.mean = np.mean(x, axis=0)
      x = x - self.mean

      cov = np.cov(x.T, ddof= 1) #By default ddof is 1.
      eigen_values, eigen_vectors = np.linalg.eig(cov)
      eigen_sorted_idxs = np.argsort(eigen_values)
      eigen_sorted_idxs_reverse = eigen_sorted_idxs[ : : -1] # revers the order of the array ([ start : end : gap])
      eigen_vectors = eigen_vectors[ : , eigen_sorted_idxs_reverse]
      projected_data = np.dot(x, eigen_vectors.T)
      return projected_data[ : , : self.num_components]
```
